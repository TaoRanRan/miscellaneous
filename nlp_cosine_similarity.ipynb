```python
# Load packages
# pip install spacy
import spacy
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
# run this in Anaconda prompt for installation: python -m spacy download en_core_web_md
# Load the downloaded word embeddings:
nlp = spacy.load("en_core_web_md")

# Each word comes with a 300-dimension vector.
# Think of it like 300 different scores for each word, all in different categories.

sentences = [
    'Toyota',
    'Audi',
    'Volvo',
    'Volkswagen',
    'BMW',
    'Tesla',
    'Skoda',
    'Ford'
]

vectors = [nlp(sentence).vector for sentence in sentences]

# Compute similarities
similarities = cosine_similarity(vectors)

# Turn into a dataframe
pd.DataFrame(similarities,
            index=sentences,
            columns=sentences) \
            .style \
            .background_gradient(axis=None)
```
